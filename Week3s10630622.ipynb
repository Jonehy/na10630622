We explore the definite integration of function $f(x)=x^2$:

$$
\begin{align*} 
I  &=\int_{0}^{1} f(x) dx \\
   &=\tfrac{1}{3} x^3 
\Big|_0^1=\tfrac{1}{3} 1^\3 -0=\tfrac{1}{3}
\end{align*}
$$

The result I=1/3.

```python
from scipy.integrate import quad
def integrand(x, a, b):
    return a*x**2+b
a = 1
b=0
I = quad(integrand, 0, 1, args=(a,b))
I
```
