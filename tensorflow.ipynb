```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
import tensorflow as tf


def f(x, y):
    return X*X+Y*Y
    #return np.sin(1/2*X*X-1/4*Y*Y+3)*np.cos(2*X+1-np.exp(Y))
    
x = np.linspace(-1, 1, 30)
y = np.linspace(-1, 1, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

def g(x,y):
  return x*x+y*y

#Add a color bar which maps values to colors
fig=plt.figure()
ax = plt.axes(projection='3d')

# Customize the z axis.
ax.set_xlim(-1.01, 1.01)
ax.set_ylim(-1.01, 1.01)
ax.set_zlim(-0.45, 0.45)

# Plot the surface.
ax = plt.axes(projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

ax.set_title('surface')
ax.view_init(30,135)

# Add a color bar which maps values to colors.
ax.xaxis.set_ticks(np.arange(-1, 1.5, 0.5))
ax.yaxis.set_ticks(np.arange(-1, 1.5, 0.5))
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```
x=tf.Variable(tf.random_uniform([1], -10, 10.0))
y=tf.Variable(tf.random_uniform([1], -10, 10.0))
```

```
# 1.4 Build training graph
loss=g(x,y)
optimizer=tf.train.GradientDescentOptimizer(0.5)
train=optimizer.minimize(loss)
init=tf.initialize_all_variables()
```

```
# 1.5 Create a session and launch the graph.
sess=tf.Session()
sess.run(init)
print(sess.run([x,y]))
```

```
for step in range(201):
    sess.run(train)
    print(sess.run([x,y]))
```
